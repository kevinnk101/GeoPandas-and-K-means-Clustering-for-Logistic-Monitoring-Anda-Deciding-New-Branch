```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler
```

```python
minmaxscaler = MinMaxScaler()
X = minmaxscaler.fit_transform(df_grupmap[['longitude','latitude']])
```

```python
inertias = []
sil_score = []
for k in range(2,21):
  km = KMeans(n_clusters=k).fit(X)
  inertias.append(km.inertia_)
  sil_score.append(silhouette_score(X,km.labels_))
```

```python
plt.plot(range(2,21),inertias)
plt.xlabel('K')
plt.ylabel('Inertia')
```

```python
plt.plot(range(2,21),sil_score)
plt.xlabel('K')
plt.ylabel('silhoette score')
```

```python
silhouette_scores = []
for k in range(2, 21):
    km = KMeans(n_clusters=k)
    km.fit(X)
    silhouette_scores.append(silhouette_score(X, km.labels_))

print("Silhouette scores:", silhouette_scores)
```

```python
sil_score = [0.9082855642468796, 0.9119055458789157, 0.8222111380845125, 0.8228151155901025, 0.824668585616767, 0.8247255685641967, 0.8352439775850427, 0.8273029019461963, 0.8263816212863194, 0.8264477821743723, 0.8325946589755432, 0.5083691960327325, 0.5153547480135025, 0.5033355220331611, 0.5131347982261385, 0.5078499098435703, 0.49217501123600127, 0.43047075941764185, 0.4667953262183304]

best_score = max(sil_score)
best_k = sil_score.index(best_score) + 2

print("The best silhouette score is: ", best_score)
print("The corresponding value of K is: ", best_k)
```

```python
import seaborn as sns

def cluster_map_plot(k):
  km = KMeans(n_clusters=k).fit(X)
  minmaxscaler.inverse_transform(km.cluster_centers_)
  color_dict={0:'maroon',1:'red',2:'blue',3:'cyan',4:'black',5:'snow',6:'yellow',7:'palegreen',8:'teal',9:'orange',10:'slategrey',11:'hotpink',12:'darkviolet',13:'cornsilk'}
  fig,ax = plt.subplots(figsize=(16,8))
  df_sumsel['geometry'].plot(ax=ax)
  plt.axvline(x=102, color='red', linestyle='-')
  plt.axvline(x=106.2, color='red', linestyle='-')
  plt.axhline(y=-1.5, color='blue', linestyle='-')
  plt.axhline(y=-5.0, color='blue', linestyle='-')

  plt.xlim(102, 106.2)
  plt.ylim(-5.0, -1.5)

  sns.scatterplot(x=df_grupmap['longitude'],y=df_grupmap['latitude'],hue=km.labels_,palette=color_dict)
  centers = minmaxscaler.inverse_transform(km.cluster_centers_)
  sns.scatterplot(x=centers[:,0],y=centers[:,1],s=500,marker='*')
  return centers

```

```python
c3 = cluster_map_plot(3)
```
